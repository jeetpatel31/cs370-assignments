# Segment Anything Model Finetuning for Remote Sensing Applications

##### Milestone 1: Get to know Geographical Information Systems (GIS) and set up your development environment See this tutorial that is best to be followed in a desktop installation of QGIS. 

#### Use the example docker container if you have access to a local NVIDIA GPU. Note that you need to do the project in Pytorch and you need to edit the devcontainer.json file to point to the right container for PyTorch. In addition to python scripts you will also need python notebooks to experiment and visualize the SAM model outputs as well as geospatial data. The suggested container ships with Jupyter and if you use Google Colab you are obviously using notebooks.
